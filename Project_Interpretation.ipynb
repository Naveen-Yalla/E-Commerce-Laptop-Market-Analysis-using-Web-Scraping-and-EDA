# Laptop Dataset - Project Interpretation & Conclusion

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Cleaned_Laptops_Data.csv')

print('Rows, Columns:', df.shape)
df.head()
```

```python
df.info()
display(df.describe(include='all').T)
```

## Dataset Interpretation
- **Total Entries:** 774 laptops.  
- **Columns present:** 9 - `Brand_name`, `Device_name`, `Operating system`, `Ram(in GB)`, `Storage(in GB)`, `Price(in ₹)`, `Discount(in percentages)`, `Rating`, `Warrenty`.  
- **Data types:** mixture of categorical and numerical.  
- **Quick notes:** No nulls detected in the cleaned dataset; many unique models and multiple brands represented.


```python
brand_counts = df['Brand_name'].value_counts()
brand_counts_head = brand_counts.head(10)

print('Top 10 brands by listing count:')
print(brand_counts_head)

plt.figure(figsize=(10,5))
brand_counts_head.plot(kind='bar')
plt.title('Top 10 Brands by Number of Listings')
plt.xlabel('Brand Name')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## Hardware Insights
- **Common RAM sizes:** 8 GB and 16 GB dominate the dataset, with average RAM around ~12 GB.  
- **Storage:** 512 GB SSD is the most frequent storage option; average storage is close to 512 GB.  
- **Implication:** Market focus on mid-range to upper-mid-range laptops with SSDs and balanced RAM configurations.


```python
ram_summary = df['Ram(in GB)'].value_counts().sort_index()
storage_summary = df['Storage(in GB)'].value_counts().sort_index()

print('RAM distribution:')
print(ram_summary)
print('\nStorage distribution:')
print(storage_summary)

plt.figure(figsize=(8,4))
plt.hist(df['Ram(in GB)'], bins=range(min(df['Ram(in GB)']), max(df['Ram(in GB)'])+4,4))
plt.title('Distribution of RAM (in GB)')
plt.xlabel('RAM (GB)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,4))
plt.hist(df['Storage(in GB)'], bins=10)
plt.title('Distribution of Storage (in GB)')
plt.xlabel('Storage (GB)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Pricing Trends
- **Price range:** From around ₹15,000 to premium devices exceeding ₹200,000.  
- **Average (approx):** Mid-market around ₹52,000 (this value can be computed exactly in the notebook).  
- **Discounts:** Large discounts (30–50%) are common, indicating active e-commerce promotions.


```python
price_stats = df['Price(in ₹)'].describe()
discount_stats = df['Discount(in percentages)'].describe()

print('Price summary:\n', price_stats)
print('\nDiscount summary:\n', discount_stats)

plt.figure(figsize=(8,4))
plt.hist(df['Price(in ₹)'], bins=30)
plt.title('Price Distribution (₹)')
plt.xlabel('Price (₹)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## Rating Insights
- **Average rating:** ~4.2/5 — high overall satisfaction.  
- **Observation:** Mid-range laptops (approx ₹40k–70k) often show better ratings, indicating value-for-money models perform well with users.


```python
rating_stats = df['Rating'].describe()
print('Rating summary:\n', rating_stats)

plt.figure(figsize=(8,5))
plt.scatter(df['Price(in ₹)'], df['Rating'], alpha=0.6)
plt.title('Price vs Rating')
plt.xlabel('Price (₹)')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

## Operating System & Warranty
- **Operating Systems:** Predominantly Windows 11 Home. Other OS entries (macOS, ChromeOS) are limited.  
- **Warranty:** Mostly 1-year manufacturer or domestic warranty; some listings include onsite or extended warranties (2–3 years). Warranty offerings can influence purchase decisions.


## Final Conclusions
1. The dataset represents a **Windows-dominated, e-commerce-driven laptop market** with a focus on **8-16 GB RAM** and **512 GB SSD** configurations.  
2. **Pricing** centers around the mid-market (~₹50k), with aggressive discounts commonly applied.  
3. **User ratings** are generally high (avg ~4.2), and value for money models in the mid-price range tend to perform better.  
4. **Business insight:** For brands and sellers - prioritize well-balanced configurations (RAM+SSD) and strategic discounts; extended warranty options can act as differentiators.  
5. **Next steps:** deeper brand-level comparative analysis, time-series pricing trends, feature engineering (CPU, GPU categories if available), and predictive modeling for price or rating.
