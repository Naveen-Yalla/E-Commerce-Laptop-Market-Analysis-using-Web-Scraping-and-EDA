```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
Price =[]
Rating = []
Discount = []
Device_name = []
Brand_name = []
Storage = []
Ram = []
OS = []
Warranties = []
Processors = []

for i in range(1,41):
    url = f"https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text)

    
    for i in soup.find_all("div",class_="tUxRFH"):    #Container
        r = i.find("div",class_="Nx9bqj _4b5DiR")    #Price
        if r:
            Price.append(r.text)
        else:
            Price.append(np.nan)


        r = i.find("div",class_="XQDdHH")              #ratings
        if r:
            Rating.append(r.text)
        else:
            Rating.append(np.nan)


        r = i.find("div",class_="UkUFwK")              #Discount
        if r:
            Discount.append(r.text)
        else:
            Discount.append(np.nan)



        
        text = i.get_text(strip=True)                 #Device name
        # Regex to capture text before " - (" pattern
        match = re.findall(r"Add to Compare(\w+.+)\s\-\s\(", text)
        if match:
            Device_name.append(match[0])
        else:
            Device_name.append(np.nan)


        text = i.get_text(strip=True)                 #Brand name
        match = re.findall(r"^(?:Add to Compare)?(ASUS|Acer|CHUWI|DELL|HP|Lenovo|MSI|Samsung|Ultimus|Infinix|Apple|Primebook|GIGABYTE|MOTOROLA|ZEBRONICS)", text)
        if match:
            Brand_name.append(match[0])
        else:
            Brand_name.append(np.nan)


        text = i.get_text(strip=True)                 #Ram name
        
        match = re.findall(r"Add to Compare\w+.+\s\-\s\((\d+)\s?GB\/", text)
        if match:
            Ram.append(match[0])
        else:
            Ram.append(np.nan)



        text = i.get_text(strip=True)                    # Storage             
        match = re.findall(r"Add to Compare\w+.+\s\-\s\(\d+\s?GB\/(\d+)\s?GB", text)
        if match:
            Storage.append(match[0])
        else:
            Storage.append(np.nan)



        text = i.get_text(strip=True)                               
        match = re.findall(r"(?:Windows\s\d{1,2}(?:\s(?:Home|Pro))?|DOS)", text) #OS
        if match:
            OS.append(match[0])
        else:
            OS.append(np.nan)



    for i in soup.find_all("ul",class_="G4BRas"): 
        text = i.text.strip()                                    # Regex pattern to match warranty information
        warranty_match = re.search(r'\d+\s*(?:Year|Years|Month|Months).*?Warrant[iy].*', text, re.IGNORECASE)
        if warranty_match:
            Warranties.append(warranty_match.group())
        else:
            Warranties.append(np.nan)


   
        text = i.text.strip()
        processor_match = re.search(                                                       # Regex pattern to match processor information
        r'(Intel Core (?:i[3579]|Ultra \d+) Processor \(\d+(?:st|nd|rd|th) Gen\)|'
        r'AMD Ryzen [3579] (?:Quad|Hexa|Octa) Core Processor|'
        r'Qualcomm Snapdragon X Elite Processor)',
        text,
        re.IGNORECASE)
        if processor_match:
            Processors.append(processor_match.group())
        else:
            Processors.append(np.nan)






```

```python
dict = {
    "Brand_name":Brand_name,
    "Device_name": Device_name,
    "Operating system": OS,
    "Ram(in GB)": Ram,
    "Storage(in GB)": Storage,
    "Price(in ₹)": Price,
    "Discount(in percentages)": Discount,
    "Rating": Rating,
    "Warrenty":Warranties,
    "Processor":Processors
}
```

```python
df = pd.DataFrame(dict)
df
```

```python
df.to_csv("Laptops_data_from_flipcart.csv", index=False)
```
